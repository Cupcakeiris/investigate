```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore

# Assuming 'data.csv' is the name of your CSV file
df = pd.read_csv('Data/inva.csv', sep=';')

# Create a single set of axes for all boxplots
melted_df = pd.melt(df, id_vars=['Region'], var_name='Year', value_name='Value')

# Clean and convert values to numeric, handling missing values
melted_df['Value'] = pd.to_numeric(melted_df['Value'], errors='coerce')

# Apply z-score normalization to the cleaned values
melted_df['Z_Score'] = melted_df.groupby('Region')['Value'].transform(zscore)

# Create a single set of axes for all boxplots
fig, ax = plt.subplots(figsize=(12, 8))

# Create a boxplot for each region with z-score normalized values
sns.boxplot(x='Region', y='Z_Score', data=melted_df, ax=ax)

# Set the plot title
plt.title('Boxplots for Regions (z-score normalized)')

# Show the plot
plt.show()
```
